```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}


def holder():
#Пользователь по команде "p" может узнать владельца документа по его номеру
    doc_number = input('Введите номер документа:')
    a =0
    for element in documents:
        if element['number'] == doc_number:
            print('Результат:')
            print('Владелец документа: ', element['name'])
            a = 1 
            break
    if a ==0:
        print('Результат:')
        print('Документ не найден в базе')


def n_shelf(n_doc):
    for key,value in directories.items():
        for element in value:
                if element == n_doc:
                    return key 
                    break 
    

def storage():
# Пользователь по команде “s” может по номеру документа узнать на какой полке он хранится 
    
    doc_number = input('Введите номер документа:')
    a = n_shelf(doc_number)
    if a == None:
        print('Результат:')
        print('Документ не найден в базе')
    else:
        print('Результат:')
        print('Документ хранится на полке: ', a)
                     
        
def all_doc():
# Пользователь по команде “l” может увидеть полную информацию по всем документам
     for element in documents:
        print('№: ',element['number'],' тип:',element['type'],' владелец:', element['name'],'полка хранения: ', n_shelf(element['number']))

def add_shelf():
#Пользователь по команде “as” может добавить новую полку    
    shelf = input('Введите номер полки:')
    if directories.get(shelf) == None:
        print('Полка добавлена.')
        directories[shelf] = []
        print('Текущий перечень полок: ', directories.keys())
    else:    
            print('Такая полка уже существует.') 
            print('Текущий перечень полок: ', directories.keys())
            
def del_shelf():
#print('Текущий перечень полок: ', directories.keys())    
    shelf = input('Введите номер полки:')
    if directories.get(shelf) == None:
        print('Такой полки не существует.')
        print('Текущий перечень полок: ', directories.keys()) 
    elif directories[shelf] == []:
        del directories[shelf]
        print('Полка удалена.')
        print('Текущий перечень полок: ', directories.keys()) 
    else:
        print('На полке есть документа, удалите их перед удалением полки. ')
        print('Текущий перечень полок: ', directories.keys()) 
        
option = input('Введите команду:')
while option != 'q':
    if option == 'p':
        holder()
    elif option == 's':
        storage()    
    elif option == 'l':    
        all_doc()
    elif option == 'as':
        add_shelf()
    elif option == 'ds':
        del_shelf()
        
    option = input('Введите команду:')
```
