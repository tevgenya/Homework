```python
import pandas as pd
# 1 Вариант. Столбец "Class" в таблице Ratings
movie = pd.read_csv('C:/Users/etokarev/Downloads/ml-latest-small (1)/ml-latest-small/movies.csv')
rating= pd.read_csv('C:/Users/etokarev/Downloads/ml-latest-small (1)/ml-latest-small/ratings.csv')
def agg_rate(param):
    if param <= 2:
        return 'низкий рейтинг'
    elif param >= 4.5:
        return 'высокий рейтинг'
    else:
        return 'средний рейтинг'

rating['class'] = rating['rating'].apply(agg_rate)
rating.head(5)
```

```python
# 2 Вариант. Столбец "Class" в таблице Movies
rating_group=rating.groupby('movieId').mean()['rating']
res=pd.merge(movie,rating_group,on='movieId')
res['class']=res['rating'].apply(agg_rate)
res.head(5)
```

```python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
            }
words = pd.read_csv('C:/Users/etokarev/Downloads/ml-latest-small (1)/ml-latest-small/keywords.csv')

def reg(param):
    for key,value in geo_data.items():
        for city in value:
            if city in param:
                return key
    return 'undefined'        
            
words['region']=words['keyword'].apply(reg)

words[words['keyword'].str.contains('псков')].head(5)
```

```python
years=[i for i in range(1950, 2011)]

def production_year(row):
    for i in years:
        if str(i) in row['title']:
            return i
    return 1900


movie['year'] = movie.apply(production_year, axis=1)
merg=pd.merge(movie,rating, on='movieId')
merg.groupby('year').mean()['rating'].sort_values(ascending=False).head(15)
    
```
