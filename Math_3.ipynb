```python
#Диагональная, ортогональная, но не единичная(!)
import numpy as np
A=np.array([[-1,0,0],[0,-1,0],[0,0,-1]])
At=np.transpose(A)
At.dot(A)
```

```python
# Верхнетреугольная, ортогональная and Нижнетреугольная, ортогональная
B=np.array([[1,0],[0,1]])
np.dot(B,np.transpose(B))
```

```python
#Матрица, у которой определитель = 1, и все элементы не равны 0
G=np.array([[2,3],[1,2]])
G
np.linalg.det(G)
```

```python
m = np.array([[2, 2],
              [1, 3]])
k,l=np.linalg.eig(m)
print("Собств.значения ",k)
print ("Собств. векторы ", l)
```

```python
#Найдите спектральное разложение матрицы:
m = np.array([[1, 2],
              [2,3]])
k,l=np.linalg.eig(m)
l.dot(np.diag(k)).dot(l.T)
```

```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
digits = load_digits()
image = digits.images[3]
plt.matshow(image) 
plt.show()
```

```python
x=[]
y=[]
for i in range(1,8,1):
     print(i)
     pca = PCA(n_components=i)
     pca.fit(image)
     image_new=pca.transform(image)

     plt.matshow(image_new) 
     plt.show()

     image_back=pca.inverse_transform(image_new)
     plt.matshow(image_back) 
     plt.show()

     y_true = image
     y_pred = image_back
     x.append(i)   
     y.append(mean_squared_error(y_true, y_pred))
```

```python
plt.scatter(x,y)
```
