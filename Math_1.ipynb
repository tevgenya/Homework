```python
import numpy as np
import matplotlib.pyplot as plt
import math

x=np.array([1,1])
y=np.array([2,0])
z=np.array([0,2])
m=x+y+z

plt.plot(x[0],x[1],'o')
plt.plot(y[0],y[1],'o')
plt.plot(z[0],z[1],'o')
plt.plot(m[0],m[1],'o')

c=np.degrees(math.acos(np.dot(x,y) / (np.linalg.norm(x) * np.linalg.norm(y))))
c
```
