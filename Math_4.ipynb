```python
import sympy as s
from scipy.misc import derivative

x=s.Symbol('x')
y=s.Symbol('y')
y=s.diff(x**5 + 4*s.sin(2*x)+ s.cos(3*x+3),x)
print('Первая производная ',y)
z=s.diff(y,x)
print('Вторая производная ',z)
```

```python
def f(x):
    return x**5 + 4*s.sin(2*x)+ s.cos(3*x+3)

print('Первая ', derivative(f,1, n=1))
print('Вторая ', derivative(f,1, n=2))
```
