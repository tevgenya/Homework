```python
import pandas as pd
data=pd.read_csv('C:/Users/etokarev/OneDrive - Philip Morris International/Desktop/private/Netology/Python/Homework/_23/vgsales.csv')
data.head(5)
```

```python
# H0- средний рейтинг критиков >= 80% (критики хорошо относятся к играм)
# H1  -средний рейтинг менее 80


from scipy import stats as st
data_upgrade=data['Critic_Score'].dropna()
print(data_upgrade.mean())

s,p=st.ttest_1samp(data_upgrade,80,alternative='less')

if p < 0.05:
    print('p= ',p,' Отвергаем H0, рейтинг менее 80%, критики плохо относятся к играм')
else:
    print ('p= ',p,' не отвергаем H0, критики хорошо относятся к играм') 
```

```python
#HO -критикам все равно, им одинаково нравятся стрелялки и стратегии
#H1 -критикам нравятся больше нравиятся стратегии

data_pc=data[data['Genre'] == "Shooter"]['Critic_Score'].dropna()
data_ps4=data[data['Genre']== "Strategy"]['Critic_Score'].dropna()
s,p= st.ttest_ind(data_pc, data_ps4)
if p< 0.05:
    print('p= ',p, ' при уровне значимости 5% H0 отвергаем, критикам больше нравятся стратегии')
else: 
    print ('p= ',p,' при уровне значимости 5% H0 не отвергаем,  критикам одинаково нравятся стрелялки и стратегии')
```

```python
spam_data=pd.read_csv('C:/Users/etokarev/OneDrive - Philip Morris International/Desktop/private/Netology/Python/Homework/_23/spam.csv')
spam_data.head(5)
```

```python
#1) Привидите весь текст к нижнему регистру
def data_lower(row):
    return row['Message'].lower()

spam_data['Message']=spam_data.apply(data_lower, axis=1)
spam_data.head(5)
```

```python
#2) Удалите мусорные символы;
import re
def data_only_words(row):
    return re.sub('[\W_]+',' ', row['Message'])

spam_data['Message']=spam_data.apply(data_only_words, axis=1)
spam_data.head(5)
```

```python
#conda install nltk.corpus
#import nltk
#nltk.download('stopwords')
```

```python
#3) Удалите стоп-слова;
from nltk.corpus import stopwords
stopwords_set = set(stopwords.words('english'))

def data_wo_stopwords(row):
    new_text=''
    spitted_text = row['Message'].split()
    for word in spitted_text:
        if word not in stopwords_set:
            new_text =new_text + ' '+ word
    return new_text

spam_data['Message']=spam_data.apply(data_wo_stopwords, axis=1)
spam_data.head(5)
```

```python
#nltk.download('wordnet')
```

```python
#4) Привидите все слова к нормальной форме;
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

def lemma(row):
    new_text=''
    spitted_text = row['Message'].split()
    for word in spitted_text:
        new_text = new_text+' '+ wordnet_lemmatizer.lemmatize(word)
    return new_text    
        
spam_data['Message']=spam_data.apply(lemma, axis=1)  
spam_data.head(5)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(spam_data['Message'])
names = tfidf.get_feature_names()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=names)
tfidf_matrix
```

```python
from  sklearn.model_selection import train_test_split


x = tfidf_matrix
y = spam_data['Category']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

model=LogisticRegression(random_state=42)
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score (y_test, model.predict(x_test))
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, model.predict(x_test),labels=["spam", "ham"])
```
